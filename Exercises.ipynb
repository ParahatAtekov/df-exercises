**Exercise 12**

Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
data = np.random.rand(10, 4)
df = pd.DataFrame(data, columns = ['Col_1', 'Col_2', 'Col_3', 'Col_4'])

# Convert some values to NaN
df.iloc[2, 1] = np.nan
df.iloc[4, 3] = np.nan
df.iloc[7, 2] = np.nan

# Create a style function to highlight NaN values
def highlight_nan(value):
    if pd.isna(value):
        return 'background-color: green'
    else:
        return ''

# Apply the style function to the dataframe
styled_df = df.style.applymap(highlight_nan)

# Display the styled dataframe
styled_df
```
